```python
#bfs로 풀어보기-2 (실패닷)
from collections import deque

n,m=map(int,input().split())
ice=[0 for _ in range(n)]
flag=True
#visited=[[True]*(m) for i in range(n)]
#print(visited)
for i in range(n):
    ice[i]=list(map(int,input()))


def bfs(graph,x,y):
    queue=deque([x,y])
    
    #굳이 visited를 써서 방문체크를 안해줘도 되겠다.
    
    if x<=-1 or x>=n or y<=-1 or y>=m:
        return False
        
    while queue:
        X,Y=queue.popleft() #왜 unpack하다고 나오지?
        if X<=-1 or X>=n or Y<=-1 or Y>=m:
            return False
     
        
        if graph[X-1][Y]==0:
            graph[X-1][Y]=1
            queue.append([X-1,Y])
        if graph[X][Y-1]==0:
            graph[X][Y-1]=1
            queue.append([X,Y-1])
        if graph[X+1][Y]==0:
            graph[X+1][Y]=1
            queue.append([X+1,Y])
        if graph[X][Y+1]==0:
            graph[X][Y+1]=1
            queue.append([X,Y+1])
        return True

    return False

answer=0
for i in range(n):
    for j in range(m):
        if bfs(ice,i,j)==True:
            answer+=1
print(answer)
# https://konghana01.tistory.com/69 참고하기
```
